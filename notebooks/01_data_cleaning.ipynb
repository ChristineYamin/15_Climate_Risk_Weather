This notebook prepares raw daily weather data collected from the NASA POWER API.

The goal is to clean the dataset, create useful time-based features, generate climate risk indicators, and save a processed dataset for later analysis and dashboard development.

```python
# Convert Date Column
df["date"] = pd.to_datetime(df["date"])
print("Minimum date:", df["date"].min())
print("Maximum date:", df["date"].max())
```

## Feature Engineering

```python
# Create Time-Based Features
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["month_name"] = df["date"].dt.strftime("%b")
df["day"] = df["date"].dt.day
df["day_of_year"] = df["date"].dt.dayofyear
df[["date", "year", "month", "month_name", "day", "day_of_year"]].head()
```

```python
# Create Season Feature
def assign_season(month):
    if month in [3,4,5]:
        return "Hot Season"
    elif month in [6,7,8,9,10]:
        return "Rainy Season"
    else:
        return "Cool Season"

df["season"] = df["month"].apply(assign_season)
df[['date', 'month', 'month_name', 'season']].head()
```

```python
# Create climate risk indicator columns
df["heat_risk_day"] = np.where(df["temperature_max"] >= 35,1,0)
df["very_hot_day"] = np.where(df["temperature_max"] >= 38,1,0)
df["heavy_rain_day"] = np.where(df["rainfall"] >= 50,1,0)
df["dry_day"] = np.where(df["rainfall"] < 1,1,0)

df[
    [
        "date",
        "city",
        "temperature_max",
        "rainfall",
        "heat_risk_day",
        "very_hot_day",
        "heavy_rain_day",
        "dry_day",
    ]
].head()
```

```python
# Create Temperature Range Feature
df["temperature_range"] = df["temperature_max"] - df["temperature_min"]
df[["date", "city", "temperature_max", "temperature_min", "temperature_range"]].head()
```

```python
# Create Rainfall Category
def categorize_rainfall(rainfall):
    if rainfall == 0:
        return "NO Rain"
    elif rainfall < 10:
        return "Light Rain"
    elif rainfall < 50:
        return "Moderate Rain"
    else:
        return "Heavy Rain"
df["rainfall_category"] = df["rainfall"].apply(categorize_rainfall)
df[["date", "city", "rainfall", "rainfall_category"]].head()
```

```python
# Create Temperature Category
def categorize_temperature(temp_max):
    if temp_max < 30:
        return "Normal"
    elif temp_max < 35:
        return "Warm"
    elif temp_max < 38:
        return "Hot"
    else:
        return "Extreme Heat"

df["temperature_category"] = df["temperature_max"].apply(categorize_temperature)
df[["date", "city", "temperature_max", "temperature_category"]].head()
```

```python
# Create Daily Climate Risk Score
df["daily_climate_risk_score"] = (
    df["heat_risk_day"] * 40
    + df["heavy_rain_day"] * 30
    + df["dry_day"] * 10
    + df["very_hot_day"] * 20
)

df["daily_climate_risk_score"] = df["daily_climate_risk_score"].clip(upper=100)
df[[
    "date",
    "city",
    "heat_risk_day",
    "heavy_rain_day",
    "dry_day",
    "very_hot_day",
    "daily_climate_risk_score",
]].head()
```

```python
# Create Risk level
def assign_risk_level(score):
    if score <= 30:
        return "Low"
    elif score <= 60:
        return "Moderate"
    else:
        return "High"

df["risk_level"] = df["daily_climate_risk_score"].apply(assign_risk_level)
df[["date", "city", "daily_climate_risk_score", "risk_level"]].head()
```
